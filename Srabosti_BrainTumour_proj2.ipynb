```python
for dirname, _, filenames in os.walk('E:\\bACKUP\\mi\\PROJECTS 2018 IT\\BrainTumor\\Brain Tumor Prediction\\archive'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Get the path of files
image_directory='E:\\bACKUP\\mi\\PROJECTS 2018 IT\\BrainTumor\\Brain Tumor Prediction\\archive\\'

no_tumor_images=os.listdir(image_directory+ 'no\\')
yes_tumor_images=os.listdir(image_directory+ 'yes\\')
# initialize dataset and label arrays
dataset=[]
label=[]
# set input size
INPUT_SIZE=64
```

```python
#loop over each image in each category
for i , image_name in enumerate(no_tumor_images):
    #read the image if its extension is .jpg
    if(image_name.split('.')[1]=='jpg'):
        image=cv2.imread(image_directory+'no/'+image_name)
        image=Image.fromarray(image,'RGB')
        #resize the image
        image=image.resize((INPUT_SIZE,INPUT_SIZE))
        #append image arry in dataset list and its label in label list
        dataset.append(np.array(image))
        label.append(0)

        # same for yes images
for i , image_name in enumerate(yes_tumor_images):
    if(image_name.split('.')[1]=='jpg'):
        image=cv2.imread(image_directory+'yes/'+image_name)
        image=Image.fromarray(image, 'RGB')
        image=image.resize((INPUT_SIZE,INPUT_SIZE))
        dataset.append(np.array(image))
        label.append(1)
```

```python
dataset=np.array(dataset)
dataset
```

```python
label=np.array(label)
label
```

```python
print(len(label))
```

```python
x_train, x_test, y_train, y_test = train_test_split(dataset, label, test_size = 0.2, random_state = 42)
```

```python
print(x_train.shape)
print(y_train.shape)
print("#######################################")
print(x_test.shape)
print(y_test.shape)
```

```python
# Normalise the data for training purpose
x_train = normalize( x_train, axis =1)
x_test = normalize( x_test, axis =1)

# Normalise the data for training purpose
#y_train = to_categorical(y_train, num_classes=2)
#y_test = to_categorical(y_test, num_classes=2)
```

```python
# Building a simple CNN model
model = Sequential()

model.add(Conv2D(32, (3,3), input_shape = (INPUT_SIZE, INPUT_SIZE, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32, (3,3), kernel_initializer = 'he_uniform'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3),  kernel_initializer = 'he_uniform'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
#model.add(Dense(2))
model.add(Activation('sigmoid'))
# model.add(Activation('softmax'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train,
batch_size = 16,
verbose = 1, epochs = 10,
validation_data = (x_test, y_test),
shuffle = 'False')
```

```python
model.save('Brain_Tumor_detection.h5')
```
